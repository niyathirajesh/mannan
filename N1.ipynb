```python
import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5]
sales = [100, 120, 140, 130, 160]
plt.plot(days,sales)
plt.title("Daily Sales")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
subjects = ["Maths", "Physics", "Chemistry"]
marks = [80, 92, 75]
plt.bar(subjects,marks)
plt.title("Marks in Subjects")
plt.xlabel("Subjects")
plt.ylabel("marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
subjects = ["Maths", "Physics", "Chemistry"]
marks = [80, 92, 75]
plt.barh(subjects,marks)
plt.title("Marks in Subjects")
plt.xlabel("Subjects")
plt.ylabel("marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
marks = [40, 30, 20, 10]
subjects = ["Maths", "Physics", "Chemistry", "English"]
plt.pie(marks, labels = subjects)
plt.show()
```

```python
import matplotlib.pyplot as plt
hours = [1, 2, 3, 4, 5]
marks = [40, 50, 60, 70, 80]
plt.scatter(hours,marks)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.show()
```
